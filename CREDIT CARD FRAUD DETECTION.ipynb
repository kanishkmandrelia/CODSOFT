Importing Libraries

Making 2 different datasets on basis of "0"(legit) and "1"(fraud).

```python
df_legit = df[df["Class"]==0]
df_fraud = df[df["Class"]==1]
```

```python
df_legit["Amount"].describe()
```

```python
df_fraud["Amount"].describe()
```

```python
df.groupby("Class").mean()
```

Making a new dataset which containsequal number of class as the "fraud" count is '492' and "legit" count is '284315'.

```python
legit_sample = df_legit.sample(n= 492)
```

```python
new_sample = pd.concat([legit_sample, df_fraud], axis=0)
```

This is our new datset which will going to use further for appyling models.

```python
new_sample
```

```python
new_sample["Class"].value_counts()
```

```python
new_sample.groupby("Class").mean()
```

Spliting the new datset into train and test. 

```python
x = new_sample.drop(columns=["Class"], axis= 1)
y = new_sample.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

Now, will apply the model i.e; Logistic Regression.

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

Will see the accuracy of both: 'train' and 'test' data; to see whether the model is "proper" or "overfitting" or "underfitting".

```python
train_pred = lr.predict(x_train)
train_acc = accuracy_score(train_pred, y_train)
train_acc
```

```python
test_pred = lr.predict(x_test)
test_acc = accuracy_score(test_pred, y_test)
test_acc
```

The model is "proper".
